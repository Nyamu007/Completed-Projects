```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pages = [x for x in range(1, 51)]  # Looping over 50 pages
product_name = []
new_price = []
former_price = []
disc_perc = []

for page in pages:
    url = f'https://www.jumia.co.ke/catalog/?q=laptop&page={page}#catalog-listing'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')

    # Extracting product names
    names = soup.find_all('h3', class_='name')
    for name in names:
        product_name.append(name.text.strip())

    # Extracting current prices
    current_prices = soup.find_all('div', class_='prc')
    for price in current_prices:
        new_price.append(price.text.strip())

    # Extracting old prices
    old_prices = soup.find_all('div', class_='old')
    for price in old_prices:
        former_price.append(price.text.strip())

    # Extracting discount percentages
    discounts = soup.find_all('div', class_='bdg _dsct _sm')
    for dis in discounts:
        disc_perc.append(dis.text.strip())

# Ensure all lists have the same length by filling missing entries with None
max_length = max(len(product_name), len(new_price), len(former_price), len(disc_perc))

product_name.extend([None] * (max_length - len(product_name)))
new_price.extend([None] * (max_length - len(new_price)))
former_price.extend([None] * (max_length - len(former_price)))
disc_perc.extend([None] * (max_length - len(disc_perc)))

# Creating a DataFrame
data = {
    'Name': product_name,
    'new_price': new_price,
    'former_price': former_price,
    'disc_perc': disc_perc
}

laptop = pd.DataFrame(data)
laptop.shape  # Should give you the shape of the DataFrame

# Display the first few rows of the DataFrame to verify
laptop.head()
```
